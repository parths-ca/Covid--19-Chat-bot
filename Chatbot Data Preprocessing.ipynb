### Import Required Packages

```python
import pandas as pd
from IPython.display import display
```

### Read all Datasets

```python
health_canada = pd.read_json(path_or_buf = 'HealthCanadaFAQ.jsonl', lines=True)
cdc = pd.read_csv('covid_faq.csv')
un = pd.read_csv('COVID19_FAQ.csv')
```

```python
display(health_canada.head(10))
display(cdc.head(10))
display(un.head(10))
```

### 1. Copy Questions and Answers from Health Canada and Rename Columns for easier concatenation

```python
health_canada = health_canada[['questionText','answerText']].copy()
health_canada.rename(inplace=True,columns={'questionText':'questions','answerText':'answers'})
display(health_canada.head(10))
```

### 2. Concatenate the Dataframes to save one master CSV

```python
final_df = pd.concat([health_canada,cdc,un],ignore_index = True)
display(final_df.head(10))
```

```python
final_df.dropna(inplace=True)
final_df = final_df[final_df.answers != ""].copy()
```

```python
final_df.to_csv("FAQData.csv",index= False)
```
